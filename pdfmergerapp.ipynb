## Implement PDF Upload

### Subtask:
Add functionality to the Flask app to handle multiple PDF file uploads from the user through the web interface.


**Reasoning**:
First, I need to modify the `templates/index.html` file to include the file upload form as specified in the instructions. This will allow users to select and upload multiple PDF files.



```python
import os

# Create the templates directory if it doesn't exist (already done, but ensuring robustness)
os.makedirs('templates', exist_ok=True)

# Modify the index.html file to include the upload form
with open('templates/index.html', 'w') as f:
    f.write("""<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>PDF Merger</title>
</head>
<body>
    <h1>PDF Merger</h1>
    <form action="/upload" method="POST" enctype="multipart/form-data">
        <input type="file" name="pdf_files" multiple accept=".pdf">
        <input type="submit" value="Upload and Merge PDFs">
    </form>
</body>
</html>
""")

print("Modified templates/index.html to include the upload form.")
```

**Reasoning**:
Now that the HTML form is in place, the next step is to modify the `app.py` file to handle the file uploads, including importing `secure_filename`, creating the `/upload` route, iterating and validating uploaded files, saving them, and redirecting the user as specified in the instructions.



```python
from flask import Flask, render_template, request, redirect, url_for
from werkzeug.utils import secure_filename
import os

app = Flask(__name__)
app.config['UPLOAD_FOLDER'] = 'uploads'

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/upload', methods=['POST'])
def upload_files():
    if 'pdf_files' not in request.files:
        return redirect(url_for('index'))

    files = request.files.getlist('pdf_files')
    uploaded_count = 0
    for file in files:
        if file.filename == '':
            continue
        if file and file.filename.lower().endswith('.pdf'):
            filename = secure_filename(file.filename)
            file.save(os.path.join(app.config['UPLOAD_FOLDER'], filename))
            uploaded_count += 1

    # In a real app, you might want to show a success message or list uploaded files
    # For this subtask, we just redirect.
    print(f"Successfully uploaded {uploaded_count} PDF files.")
    return redirect(url_for('index'))

if __name__ == '__main__':
    os.makedirs(app.config['UPLOAD_FOLDER'], exist_ok=True)
    app.run(debug=True)
```
